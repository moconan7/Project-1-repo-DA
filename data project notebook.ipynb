```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='/content/drive/MyDrive/Data/Data analysis Project.csv'
df=pd.read_csv(path)
```

```python
# Visualizations
# 1. Distribution of work hours
plt.figure(figsize=(8, 6))
sns.histplot(df['On average, how many hours do you work per week?'].dropna(), bins=10, kde=True)
plt.title('Distribution of Work Hours per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Balance rating vs. work hours
plt.figure(figsize=(8, 6))
sns.scatterplot(x='On average, how many hours do you work per week?', y='How would you rate your ability to balance work, studies, and personal life?', data=df)
plt.title('Work Hours vs. Balance Rating')
plt.xlabel('Hours per Week')
plt.ylabel('Balance Rating')
plt.show()
```

```python
# 3. Academic performance affected by work commitments
plt.figure(figsize=(8, 6))
sns.countplot(x='Do you feel your academic performance is affected by your work commitments?', data=df)
plt.title('Academic Performance Affected by Work Commitments')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Leisure time availability
plt.figure(figsize=(8, 6))
sns.countplot(x='Do you feel you have enough time for leisure activities?', data=df)
plt.title('Availability of Leisure Time')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```
